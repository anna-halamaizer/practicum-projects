# E-Commerce: Product Range Analysis

## Project description

We (a large retail chain) have recently acquired a small local store.

We wanna keep it as business, but improve the efficiency of sales. For that we need to know, what the store has been selling, what's the assortment and how the sales are going.

Ideally, in the future we want to optimize the PR (Product Range) - keep only popular and "effective" items. For now we want to understand the general picture, as well as what we need to have in stock a lot, what we do not buy and sell anymore and so on.

**Task:**

The project main business goal  is to study the assortment of the store :
- what do we sell and how it's sold
- what are popular items? What are items are not popular?
- what items generate the biggest revenue? What items are sold together and etc.

Based on our research as a team we need to make recommendations to the head managers on  how to change/improve product range.

**Description of the data:**

The dataset contains the transaction history of an on-line store that sells household goods.
The file `ecommerce_dataset_us.csv` contains the following columns:
- `InvoiceNo` — order identifier
- `StockCode` — item identifier
- `Description` — item name
- `Quantity`
- `InvoiceDate` — order date
- `UnitPrice` — price per item
- `CustomerID`

## Categorization

### Categorization by price

```python
pd.qcut(data['UnitPrice'], 5)
```

A lot of units has price up to 3 USD. I will make a groups with interval of 1 USD for them. The rest data will be divided into groups of with price of 3 to 5 USD , 5 to 13 USD and over 13 USD (expensive items, outliers):

```python
def price_group(row):
    
    if row['UnitPrice'] < 1:
        return '0 to 1'
    if row['UnitPrice'] < 2:
        return '1 to 2'
    if row['UnitPrice'] < 3:
        return '2 to 3'
    if row['UnitPrice'] < 5:
        return '3 to 5'
    if row['UnitPrice'] < 13:
        return '5 to 13'
    if row['UnitPrice'] > 13:
        return 'over 13'
```

```python
data['PriceGroup']=data.apply(price_group, axis=1)
```

```python
data.sample(5)
```

### RFM

```python
data['Date'].max()
```

```python
last_date=dt.datetime(2019,11,30)
last_date
```

```python
rfm=data.groupby('Description').agg({
    'Date': lambda x: (last_date-x.max()).days,
    'InvoiceNo': 'nunique',
    'Revenue':'sum'
}).reset_index()
rfm.rename(columns={'Date':'recency',
                    'InvoiceNo':'frequency',
                    'Revenue':'monetary'},inplace=True)
rfm.head()
```

```python
rfm.info()
```

Assigning codes to each unit. 1 - is the most recent, frequent and valuable and 4 - is the oldest, rare and cheap. 

```python
rfm['R']=pd.qcut(rfm.recency,5,[1,2,3,4],duplicates = 'drop')
rfm['F']=pd.qcut(rfm.frequency,4,[4,3,2,1])
rfm['M']=pd.qcut(rfm.monetary,4,[4,3,2,1])
rfm.head()
```

```python
rfm['RFM_code']=rfm['R'].astype(str)+rfm['F'].astype(str)+rfm['M'].astype(str)
rfm.head()
```

Adding name to each number in the code:

```python
def rfm_name(row):
    
    if row['R'] == 1:
        r_name = 'recent'
    if row['R'] == 2:
        r_name = 'less recent'
    if row['R'] == 3:
        r_name = 'old'
    if row['R'] == 4:
        r_name = 'oldest'
    
    if row['F'] == 1:
        f_name = 'top seller'
    if row['F'] == 2:
        f_name = 'popular'
    if row['F'] == 3:
        f_name = 'unpopular'
    if row['F'] == 4:
        f_name = 'rare'
        
    if row['M'] == 1:
        m_name = 'best value'
    if row['M'] == 2:
        m_name = 'good value'
    if row['M'] == 3:
        m_name = 'low value'
    if row['M'] == 4:
        m_name = 'cheap'
    
    return r_name + " / " + f_name + " / " + m_name
 
```

```python
rfm['RFM_segment'] = rfm.apply(rfm_name, axis=1)
rfm.head()
```

```python
rfm['RFM_segment'].nunique()
```

```python
data=data.merge(right=rfm[['Description','R','F','M','RFM_code','RFM_segment']], on='Description', how='left')
data.head()
```

### Key words

#### Lemmatization

```python
m = Mystem()
wordnet_lemma = WordNetLemmatizer()
```

```python
#preparing list of all lemmas
lemmas_list_all = []

for Description in data.Description.unique():
    words = nltk.word_tokenize(Description)
    lemmas = [wordnet_lemma.lemmatize(w, pos = 'n') for w in words]
    for i in lemmas:
        lemmas_list_all.append(i)
```

```python
lemmas_list_all[0:10]
```

I will select only nouns from lemmas to use them as a keyword to understand the nature of item.

```python
# Use the pos_tag() function to tag the words as nouns, verbs, etc.
lemmas_list_all = [str(x) for x in lemmas_list_all] #convert to srtings
lemmas_list_all = [x.lower() for x in lemmas_list_all] #convert to lowercase for nltk
tagged_list = nltk.pos_tag(lemmas_list_all)

# Use a list comprehension to select the nouns
nouns = [word for word, tag in tagged_list if tag == 'NN']
```

```python
nouns[0:10]
```

#### Hypothesis testing

Let's use the z-criterion to check the statistical difference between the proportions in price groups.

Hypothesis: proportions of revenue split between price groups were the same in Sep, Oct and Nov 2019

- The null hypothesis (H0) for the test is that the proportions are the same.
- The alternate hypothesis (H1) is that the proportions are not the same.

```python
def check_hypothesis(group_rev_m1,  total_rev_m1, group_rev_m2, total_rev_m2, alpha):
    p1 = group_rev_m1/total_rev_m1
    p2 = group_rev_m2/total_rev_m2
    p_combined = (group_rev_m1 + group_rev_m2) / (total_rev_m1 + total_rev_m2)
  
    difference = p1 - p2
   
    z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (1/total_rev_m1 + 1/total_rev_m2))
 
    distr = stats.norm(0, 1) 

    p_value = (1 - distr.cdf(abs(z_value))) * 2

    print('p-value: ', p_value)
    
    if p_value<alpha:
        print('Reject H0')
    else:
        print('Fail to Reject H0')
```

```python
z_data=prices_month_pivot[['2019-09', '2019-10', '2019-11']]/1000
z_data.reset_index()
```

I set alpha at 0.05.

But as I am going to perform several statistical tests simultaneously on a single data set, I will make the Bonferroni correction in order to reduce the chances of obtaining false-positive results (type I errors) and divide the critical alpha by the number of comparisons being made (alpha = 0.05/15).

```python
print('Testing hypothesis for September and October:')
for i in range(0,5):
    print(f'For price group: {z_data.index[i]}') 
    check_hypothesis(z_data.iloc[i,0],z_data.iloc[6,0], z_data.iloc[i,1], z_data.iloc[6,1],0.05/15)
    print('')
```

```python
print('Testing hypothesis for September and October:')
for i in range(0,5):
    print(f'for price group: {z_data.index[i]}') 
    check_hypothesis(z_data.iloc[i,1],z_data.iloc[6,1], z_data.iloc[i,2], z_data.iloc[6,2],0.05/15)
    print('')
```

```python
print('Testing hypothesis for September and October:')
for i in range(0,5):
    print(f'for price group: {z_data.index[i]}') 
    check_hypothesis(z_data.iloc[i,0],z_data.iloc[6,0], z_data.iloc[i,2], z_data.iloc[6,2],0.05/15)
    print('')
```

**Conclusion:**

Based on z-test of proportions, we can conclude, that statistically proportions of revenue split between price groups were the same in Sep, Oct and Nov 2019.
